```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import time

import seaborn as sns
sns.set()
```

```python
phases = {}
G = {}
for i in range(1,12): 
  var_name = "phase" + str(i)
  file_name = "https://raw.githubusercontent.com/ragini30/Networks-Homework/main/" + var_name + ".csv"
  phases[i] = pd.read_csv(file_name, index_col = ["players"])
  phases[i].columns = "n" + phases[i].columns
  phases[i].index = phases[i].columns
  phases[i][phases[i] > 0] = 1
  G[i] = nx.from_pandas_adjacency(phases[i])
  G[i].name = var_name
```

```python
def network_size(graph):
  """Function to determine shape of a graph"""
  return f'{graph.number_of_nodes()} nodes and {graph.number_of_edges()} edges'
```

```python
def centrality_frame(graph, phase, c):
  """Function to data frame the centrality of a graph [[[Smaller Version]]].
  The normalized centrality for each of the 23 players under investigation in the following wiretapping/sizing phases."""
  if c==1:
    """degree centrality"""
    return pd.DataFrame(nx.degree_centrality(graph[phase]), index=('degree_centrality', ), columns=graph[phase].nodes())
  elif c==2:
    return pd.DataFrame(nx.betweenness_centrality(graph[phase], normalized=True), index=('betweeness_centrality', ), columns=graph[phase].nodes())
  elif c==3:
    return pd.DataFrame(nx.eigenvector_centrality(graph[phase]), index=('eigenvector_centrality', ), columns=graph[phase].nodes())
```

```python
''' Creating a node string list'''
G_nodes = ['n' + str(i) for i in range(1, 111)]
```

```python
def dataframe(graph, c):
  """Function to determine the centrality of a graph [[[Bigger Version]]].
  The normalized centrality for each of the 23 players under investigation in the following wiretapping/sizing phases.
  Returns result as dataframe"""

  D1 = pd.DataFrame(0, index=list(range(1, 12)), columns=G_nodes) #data frame filled with only 0 
  D2 = pd.DataFrame(0, index=list(range(1, 12)), columns=G_nodes)
  D3 = pd.DataFrame(0, index=list(range(1, 12)), columns=G_nodes)

  if c==1:
    for p1 in range(1, 12):
      D1.loc[p1, :]=nx.degree_centrality(G[p1])
      
  elif c==2:
    for p2 in range(1, 12):
      D2.loc[p2, :]=nx.betweenness_centrality(G[p2], normalized=True)
           
  elif c==3:
    for p3 in range(1, 12):
      D3.loc[p3, :]=nx.eigenvector_centrality(G[p3])

  if c==1:
    return D1.replace(np.nan, 0)
  elif c==2:
    return D2.replace(np.nan, 0)
  elif c==3:
    return D3.replace(np.nan, 0) 

```

```python
G_ = {}
phases_ = {}
for i in range(1,12): 
  var_name_ = "phase" + str(i)
  file_name_ = "https://raw.githubusercontent.com/ragini30/Networks-Homework/main/" + var_name_ + ".csv"
  phases_[i] = pd.read_csv(file_name_, index_col = ["players"])
  phases_[i].columns = "n" + phases_[i].columns
  phases_[i].index = phases_[i].columns
  phases_[i][phases_[i] > 0] = 1
  G_[i] = nx.from_pandas_adjacency(phases_[i], create_using=nx.DiGraph())
  G_[i].name = var_name_
```

```python
def dataframe_directed(graph, c):
  """Function to determine the centrality of a graph [[[Bigger Version]]].
  The Hubs and Authority for each of the players under investigation in the following wiretapping/sizing phases.
  Returns result as dataframe"""

  D1 = pd.DataFrame(0, index=list(range(1, 12)), columns=G_nodes) #data frame filled with only 0 
  D2 = pd.DataFrame(0, index=list(range(1, 12)), columns=G_nodes)
  # D3 = pd.DataFrame(0, index=list(range(1, 12)), columns=G_nodes)

  if c==1:
    for p1 in range(1, 12):
      D1.loc[p1, :]=nx.algorithms.link_analysis.hits( G_[1], max_iter=1000000)[0]
      
  elif c==2:
    for p2 in range(1, 12):
      D2.loc[p2, :]=nx.algorithms.link_analysis.hits( G_[1], max_iter=1000000)[-1]
           

  if c==1:
    return D1.replace(np.nan, 0)
  elif c==2:
    return D2.replace(np.nan, 0)


```

```python
def hits_frame(graph, phase, c):
  """Function to data frame the centrality of a graph [[[Smaller Version]]].
  The normalized centrality for each of the 23 players under investigation in the following wiretapping/sizing phases."""
  if c==1:
    return pd.DataFrame(nx.algorithms.link_analysis.hits(graph[phase], max_iter=1000000)[0], index=('HUBS', ), columns=graph[phase].nodes())[['n1', 'n3' ]]
  elif c==2:
    return pd.DataFrame(nx.algorithms.link_analysis.hits(graph[phase], max_iter=1000000)[-1], index=('AUTHORITY', ), columns=graph[phase].nodes())[['n1', 'n3']]
 
```

```python

hubs_n1 = []
hubs_n3 = []

star = 7 * '*'
for p in range(1, 12):
  'hubs over phases'
  print(f'{star} Phase {p} {star}')
  print(hits_frame(G_, p, 1))
  hubs_n1.append(float(hits_frame(G_, p, 1)['n1'].to_numpy()))
  hubs_n3.append(float(hits_frame(G_, p, 1)['n3'].to_numpy()))

print(hubs_n1)
print()
print(hubs_n3)
```

```python
x = np.arange(1, 12)
plt.plot(x, hubs_n1, 'bo-', label="n1")
plt.plot(x, hubs_n3, 'yo-', label="n3")
plt.xticks(x, x)
plt.legend()
plt.title('Hubs of N1 and N3 Across Investigative Phases')
plt.ylabel('hubs')
plt.xlabel('investigative phase')
plt.show()
```
